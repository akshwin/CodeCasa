**VISUALIZATION**

**MODEL BUILDING**

**SPLITTING DATASET**

```python
x = df.drop('Class',axis = 1)
y = df['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3 , random_state=1) 
```

**LOGISTIC REGRESSION**

```python
from sklearn import metrics 
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver="liblinear")
model.fit(x_train,y_train)
```

```python
logistic_predict = model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
logistic_score = model.score(x_test,y_test)
```

```python
corrmat = df.corr() 
fig = plt.figure(figsize = (12, 9)) 
sns.heatmap(corrmat, vmax = .8, square = True) 
plt.show()
```

**NAIVE BAYES**

```python
from sklearn.naive_bayes import GaussianNB # using Gaussian algorithm from Naive Bayes

model = GaussianNB()
model.fit(x_train,y_train)
```

```python
naive_predict= model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
naive_score = model.score(x_test,y_test)
```

```python
print("The Accuracy by Logistic Regression Algorithm  : ",logistic_score)
print("The Accuracy by Naive Bayes Algorithm  : ",naive_score)
```
